```python
#this is a bar graph

import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

x_axis = ['a', 'b', 'c']
y_axis = [2,8,10]
ind = np.arange(len(x_axis))
plt.bar(ind, y_axis)
plt.xticks(ind,x_axis)
```

```python
#scatter plot

plt.scatter(ind,y_axis)
plt.show()
```

```python
#line graph

plt.plot(ind,y_axis)
plt.show() 
```

```python
#histogram

values = [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]
plt.hist(values, bins = 3)
plt.show()
```
